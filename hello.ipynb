```python
# plot multiple things
circle((-1, 3), 3) + point((1,2)) + point([(-3, 3), (-3, 4), (-3, 5)], marker='d', rgbcolor='green', size=40)
```

```python
# plot multiple things using sum and list
sum([
  circle((-1, 3), 3),
  point((1,2)),
  point([(-3, 3), (-3, 4), (-3, 5)], marker='d', rgbcolor='green', size=40),
  line([(-2, 1), (0.5, 1)], rgbcolor='red'),
])
```

```python
# plot multiple functions
plot([2*x - 3, x^2], -3, 5)
```

```python
# define multiple functions and plot them at once
f(x) = 2*x - 3
g(x) = x^2
plot([f, g], -5, 5)
```

```python
# combine graphic elements and then call show on the parent Graphics object
g = sum([
  line([(-7, 4), (5, -12)]),
  point((-1, -4), size=20),
  circle((-1,-4), 10, color='green'),
  # note this function won't be drawn past its xmin and xmax
  plot(3/4*x - 13/4, -5, 6, rgbcolor='red', linestyle='--'),
])
g.show(xmin=-12, xmax=10, ymin=-17, ymax=8, aspect_ratio=1)
```

```python
# Solve a system of equations
var('x y z')
solve([
  2*x + y - 4*z == 28,
  x + y + z == 8,
  2*x - y - 6*z == 22
], x, y, z)
```

```python
# Print solution
solutions = solve(x^2 == 4, x)
for solution in solutions:
  print(f'{solution.lhs()} = {solution.rhs()}')
```
